```python
def the_order_of(ribbons):
    rows = ribbons.split('\n')
    columns = []
    for i in range(11):
        column = ''
        for row in rows:
            column += row[i]
        columns.append(column)
    
    full_lines = []    
    for row in rows:
        if '.' not in row:
            full_lines.append(list(set(row)))
    for column in columns:
        if '.' not in column:
            full_lines.append(list(set(column)))   
    
    order = ''
    while len(full_lines) > 0:
        full_lines.sort(key = len)
        upper_ribbon = full_lines[0][0]
        order += upper_ribbon        
        for i in full_lines:
            if upper_ribbon in i:
                i.remove(upper_ribbon)
        full_lines.pop(0)                  
          
    return order[::-1]
```
