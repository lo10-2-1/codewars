```python
def path_finder(maze):
    '''My solution'''
    if 'W' not in maze:
        return True
    else:
        maze = maze.split('\n')
        maze = [list(maze[i]) for i in range(len(maze))]
        N = len(maze)
        limit = N**2
        i = 0
        j = 0
        while limit != 0:
            if j<=N-1//2 or maze[i][j-1]=='W':
                for j in range(N):
                    if maze[i][j]!='W':
                        maze[i][j] = True
                    else:
                        j -= 1
                        break
            elif j>=N-1//2 or maze[i][j+1]=='W':
                for j in range(N-1, -1, -1):
                    if maze[i][j]!='W':
                        maze[i][j] = True
                    else:
                        j -= 1
                        break
            if i<=N-1//2 or maze[i-1][j]=='W':
                for i in range(N):
                    if maze[i][j]!='W':
                        maze[i][j] = True
                    else:
                        i -= 1
                        break
            elif i>=N-1//2 or maze[i+1][j]=='W':
                for i in range(N-1, -1, -1):
                    if maze[i][j]!='W':
                        maze[i][j] = True
                    else:
                        i -= 1
                        break
            if maze[N-1][N-1] == True:
                return True
            limit -= 1
    return False
```

```python
def path_finder(maze):
    '''Best solution from Codewars'''
    matrix = list(map(list, maze.splitlines()))
    stack, length = [[0, 0]], len(matrix)
    while len(stack):
      x, y = stack.pop()
      if matrix[x][y] == '.':
        matrix[x][y] = 'x'
        for x, y in (x, y-1), (x, y+1), (x-1, y), (x+1, y):
          if 0 <= x < length and 0 <= y < length:
            stack.append((x, y))
    return matrix[length-1][length-1] == 'x'
```
